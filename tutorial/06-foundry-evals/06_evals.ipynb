# Module 06 - Foundry evaluations

Quality, safety, and a custom tool-accuracy evaluator, logged to the Foundry
project. See [README.md](README.md).

```python
# Tutorial setup: find the repo root and make the repo code importable.
import sys
from pathlib import Path


def find_repo_root(start: Path) -> Path:
    for p in [start, *start.parents]:
        if (p / "requirements.txt").exists() and (p / "src").exists():
            return p
    return start


REPO = find_repo_root(Path.cwd())
for sub in ["src", "data", "webapp/backend"]:
    sys.path.insert(0, str(REPO / sub))
print("Repo root:", REPO)
```

## A quick local smoke run

A few cases, quality + tool-accuracy only, to see the output inline. The full run
goes to AML compute in the next cell.

```python
import subprocess

proc = subprocess.run(
    [sys.executable, str(REPO / "src" / "evals" / "run_evals.py"),
     "--mode", "local", "--max-cases", "4", "--skip-safety"],
    cwd=str(REPO), capture_output=True, text=True,
)
print(proc.stdout[-2500:])
print(proc.stderr[-800:])
```
