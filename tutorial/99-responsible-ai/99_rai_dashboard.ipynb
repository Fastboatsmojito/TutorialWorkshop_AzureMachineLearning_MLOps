# Module 99 - Responsible AI dashboard

Explanations and error analysis for the forecaster, attached to the registered
model. See [README.md](README.md).

```python
# Tutorial setup: find the repo root and make the repo code importable.
import sys
from pathlib import Path


def find_repo_root(start: Path) -> Path:
    for p in [start, *start.parents]:
        if (p / "requirements.txt").exists() and (p / "src").exists():
            return p
    return start


REPO = find_repo_root(Path.cwd())
for sub in ["src", "data", "webapp/backend"]:
    sys.path.insert(0, str(REPO / sub))
print("Repo root:", REPO)
```

## Submit the RAI pipeline

Runs on the `cpu-cluster`. It registers an MLflow model plus MLTable splits, then
wires the official Responsible AI components.

```python
import subprocess

proc = subprocess.run(
    [sys.executable, str(REPO / "src" / "pipeline" / "rai_dashboard.py"), "--no-stream"],
    cwd=str(REPO), capture_output=True, text=True,
)
print(proc.stdout[-2500:])
print(proc.stderr[-800:])
```

When the job finishes, open AML Studio, Models, `aeso-price-forecaster-rai`,
Responsible AI tab. Read the global SHAP explanations and use error analysis to
find the model's weakest regions.
