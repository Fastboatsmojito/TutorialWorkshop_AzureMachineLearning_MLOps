# Module 05 - The Foundry agent

GPT-4o calls the ML model as a tool, so every number is grounded.
See [README.md](README.md).

```python
# Tutorial setup: find the repo root and make the repo code importable.
import sys
from pathlib import Path


def find_repo_root(start: Path) -> Path:
    for p in [start, *start.parents]:
        if (p / "requirements.txt").exists() and (p / "src").exists():
            return p
    return start


REPO = find_repo_root(Path.cwd())
for sub in ["src", "data", "webapp/backend"]:
    sys.path.insert(0, str(REPO / sub))
print("Repo root:", REPO)
```

## Ask a forecasting question

`return_tool_outputs=True` also returns the raw tool results, which the
evaluations in Module 06 use as groundedness context.

```python
import agent

question = "What is the day-ahead power price forecast for tomorrow, and when does it peak?"
result = agent.run_agent([{"role": "user", "content": question}], return_tool_outputs=True)
print("REPLY:")
print(result["reply"])
print("TOOLS CALLED:", result["tool_trace"])
```

## Confirm the model, not the LLM, produced the numbers

```python
for t in result.get("tool_outputs", []):
    print(t["tool"], "->", str(t["output"])[:160], "...")
```

## Run a what-if scenario in plain English

```python
scenario_q = "Show me a cold snap: temperature down 8 degrees and demand up 10 percent."
scenario = agent.run_agent([{"role": "user", "content": scenario_q}], return_tool_outputs=True)
print(scenario["reply"])
print("TOOLS:", scenario["tool_trace"])
```
